```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
dataset = pd.read_csv('stock.csv')
df_tips = pd.read_csv('stock.csv')
df = pd.read_csv('stock.csv')
df.head(5)
```

```python
# HISTOGRAM PURCHASE PRICE
sns.set(font_scale=1.4)
df_tips['Purchase Price'].plot(kind='hist', figsize=(10, 10));
plt.xlabel("Total Bill Amount ($)", labelpad=14)
plt.ylabel("Frequency", labelpad=14)
plt.title("Histogram Purchase Price", y=1.015, fontsize=22);
```

```python
# HISTOGRAM CURRENT PRICE
sns.set(font_scale=1.4)
df_tips['Current Price'].plot(kind='hist', figsize=(10, 10));
plt.xlabel("Total Bill Amount ($)", labelpad=14)
plt.ylabel("Frequency", labelpad=14)
plt.title("Histogram Current Price", y=1.015, fontsize=22);
```

```python
# HISTOGRAM GAIN/LOSS
sns.set(font_scale=1.4)
df_tips['Gain/Loss'].plot(kind='hist', figsize=(10, 10));
plt.xlabel("Total Bill Amount ($)", labelpad=14)
plt.ylabel("Frequency", labelpad=14)
plt.title("Histogram Gain/Loss", y=1.015, fontsize=22);
```

```python
# Sorting data Top 5 Dividend/share
dividend = (dataset.groupby('Symbol')['Dividend/share']
                .sum()
                .reset_index()
                .sort_values(by='Dividend/share',ascending=False)
                .head(5))
#dataframe baru dari top 5 dividend share
dividend = dataset[(dataset['Symbol'].isin(dividend['Symbol'].to_list()))]
```

```python
plt.figure(figsize=(10,5))
plt.hist(dividend.groupby('Symbol')['Dividend/share'].nlargest(5), bins=10, stacked=True, range=(1,3), color='Purple')
plt.title('5 Teratas Dividend Share',fontsize=20, color='purple')
plt.xlabel('Dividend Share', fontsize = 10)
plt.ylabel('Frequency',fontsize = 10)
plt.xlim(xmin=0.9,xmax=2.5)
```

```python
plt.figure(figsize=(10,5))
# kelompokan berdasar Symbol yang memiliki Market_Value serta sortir top 5 tertinggi dan  urutkan dari terbesar ke terkecil
print(dividend.groupby('Symbol')['Market Value'].nlargest(5).sort_values(ascending=False))
plt.hist(dividend.groupby('Symbol')['Market Value'].nlargest(5), bins=10, stacked=True, range=(7000,15000), color='lightblue')
plt.title('5 Teratas Market Value',fontsize=20, color='blue')
plt.xlabel('Market Value', fontsize = 10)
plt.ylabel('Frequency',fontsize = 10)
plt.xlim(xmin=7000,xmax=15000)
```
